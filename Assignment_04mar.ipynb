```python
Q5. What is Distplot? Using Plotly express, plot a distplot.
```

```python
A distribution plot, also known as a Distplot, displays the variation in data distribution.
```

```python
import plotly.express as px
df = px.data.tips()
# hover_data canbe box,rug and violin
fig = px.histogram(df, x="total_bill", y="tip", color="sex", marginal="box",hover_data=df.columns)
fig.show()
```
